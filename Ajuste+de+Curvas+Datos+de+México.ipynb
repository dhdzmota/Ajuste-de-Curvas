# Dióxido de Carbono y Energía en México desde el año 2000. 

# Representación de datos y ajuste de curvas

El ajuste de curvas sirve para poder entender de una manera gráfica la relación de los datos con las variables $x$ y $y$ respectivas, en general puede fungir también para poder predecir ciertos valores ya que se aproxima una función a los datos. Para ilustrar esto con un ejemplo, se tomaron datos del Banco Mundial acerca de distintas características de México. En especifico son datos tomados desde el año 2000 hasta el 2014 acerca de:

- Emisión de Dióxido de Carbono en toneladas métricas per cápita.
- Uso de energía (en kilogramos de aceite equivalente per cápita.


```python
#Bibliotecas
import numpy as np
import matplotlib.pyplot as plt

#Listas con datos

#Años desde el 2000
tiempo=np.array([2000., 2001., 2002., 2003., 2004., 2005., 2006., 2007., 2008., 2009., 2010., 2011., 2012., 2013., 2014])

#Emisiones de Dióxido de Carbono (toneladas métricas per cápita) desde el año 2000
TonCO2=np.array([3.916478182, 4.006881616, 3.970934988, 4.16704652, 4.124551403, 4.299365235, 4.353173578, 4.296635487, 4.339644436, 4.120600766, 3.957657664, 4.067755192, 4.107686314, 4.001602493, 3.866241113])

#Uso de energía (kg. de aceite equivalente per cápita)
Energia=np.array([1474.592344, 1506.720352, 1495.442583, 1565.077651, 1562.085315, 1646.774813, 1659.516347, 1622.971122, 1580.662138, 1557.400198, 1489.637449, 1541.834073, 1586.612978, 1567.683404, 1513.259208])
```

## Emisión de dióxido de carbono a través del tiempo


```python
#Representación de los Datos:

#Gráfica de puntos de Emisión de Dióxido de Carbono desde el año 2000 hasta el 2014
plt.plot(tiempo, TonCO2, 'ob',label='Emisión CO2')
plt.legend(loc='best')
plt.xlabel("Tiempo(Años)")
plt.ylabel("Dióxido de Carbono (Toneladas)")
plt.show()
```

### Representación Lineal

```python
#Gráfica de Emisión de Dióxido de Carbono desde el año 2000 hasta el 2014
#Con un ajuste de curva polinomial lineal (mal ajuste) 

#Creación de polinomio, en este caso lineal (se representa como vector)
# debido a la estructura de los datos, hacer una aproximación lineal no es una buena aproximación.
polinomio_tiempo_TonCO2=np.polyfit(tiempo,TonCO2, 1)

#Función que toma en cuenta los valores del polinomio.
curva_tiempo_TonC02=polinomio_tiempo_TonCO2[0]*tiempo+polinomio_tiempo_TonCO2[1]

#Procedimiento para graficar
plt.plot(tiempo, Ton_CO2, 'ob',label='Emisión CO2')
plt.plot(tiempo, curva_tiempo_TonC02 , 'r',label='Ajuste lineal')
plt.legend(loc='best')
plt.xlabel("Tiempo (Años)")
plt.ylabel("Dióxido de Carbono (Toneladas)")
plt.title("Gráfica de Emisión de Dióxido de Carbono desde el año 2000 hasta el 2014")
plt.show()
```

### Representación Cuadrática

```python
#Gráfica de Emisión de Dióxido de Carbono desde el año 2000 hasta el 2014
#Con un ajuste de curva polinomial cuadrática 
polinomio_tiempo_TonCO2=np.polyfit(tiempo,TonCO2, 2)

curva_tiempo_TonC02=polinomio_tiempo_TonCO2[0]*tiempo**2+polinomio_tiempo_TonCO2[1]*tiempo+polinomio_tiempo_TonCO2[2]

plt.plot(tiempo, Ton_CO2, 'ob',label='Emisión CO2')
plt.plot(tiempo, curva_tiempo_TonC02 , 'r',label='Ajuste cuadrático')
plt.legend(loc='best')
plt.xlabel("Tiempo (Años)")
plt.ylabel("Dióxido de Carbono (Toneladas)")
plt.title("Gráfica de Emisión de Dióxido de Carbono desde el año 2000 hasta el 2014")
plt.show()

print("La ecuación que corresponde a la gráfica es: f(x)=",polinomio_tiempo_TonCO2[0],"x^2 +",polinomio_tiempo_TonCO2[1],"x +",polinomio_tiempo_TonCO2[0])
```

## Representación de orden 3, 4 y 5.

```python
#Gráfica de Emisión de Dióxido de Carbono desde el año 2000 hasta el 2014
#Con un ajuste de curva polinomial de orden 3, 4 y 5to

p3=np.polyfit(tiempo,TonCO2, 3)
curva3=p3[0]*tiempo**(3)+p3[1]*tiempo**(2)+p3[2]*tiempo**(1)+p3[3]*tiempo**(0)


p4=np.polyfit(tiempo,TonCO2, 4)
curva4=p4[0]*tiempo**(4)+p4[1]*tiempo**(3)+p4[2]*tiempo**(2)+p4[3]*tiempo**(1)+p4[4]*tiempo**(0)

p5=np.polyfit(tiempo,TonCO2, 5)
curva5=p5[0]*tiempo**(5)+p5[1]*tiempo**(4)+p5[2]*tiempo**(3)+p5[3]*tiempo**(2)+p5[4]*tiempo**(1)+p5[5]*tiempo**(0)



plt.plot(tiempo, Ton_CO2, 'ob',label='Emisión CO2')
plt.plot(tiempo, curva3 , 'r', label='Ajuste de orden 3')
plt.plot(tiempo, curva4 , 'g', label='Ajuste de orden 4')
plt.plot(tiempo, curva5 , 'y', label='Ajuste de orden 5')

plt.legend(loc='best')
plt.xlabel("Tiempo (Años)")
plt.ylabel("Dióxido de Carbono (Toneladas)")
plt.title("Gráfica de Emisión de Dióxido de Carbono desde el año 2000 hasta el 2014")
plt.show()
```

En este caso, el ajuste de orden 5to no fue agregado debido a que hubo un mensaje de Avertencia conforme al rango del polinomio. 

## Uso de energía a traves del tiempo

```python
#Gráfica de Uso de Energía desde el año 2000 hasta el 2014
#En este mismo programa ya se cuenta con una aproximación cuadrática y los datos 
polinomio_tiempo_Energia=np.polyfit(tiempo,Energia, 2)

polinomio_tiempo_Energia4=np.polyfit(tiempo,Energia, 4)
p4=polinomio_tiempo_Energia4


curva_tiempo_Energia=polinomio_tiempo_Energia[0]*tiempo**2+polinomio_tiempo_Energia[1]*tiempo+polinomio_tiempo_Energia[2]

curva4=p4[0]*tiempo**(4)+p4[1]*tiempo**(3)+p4[2]*tiempo**(2)+p4[3]*tiempo**(1)+p4[4]*tiempo**(0)


plt.plot(tiempo, Energia, 'ob',label='Energía')
plt.plot(tiempo, curva_tiempo_Energia , 'r',label='Ajuste de segundo orden')
plt.plot(tiempo,curva4,"g", label='Ajuste de cuarto orden')
plt.legend(loc='best')
plt.xlabel("Tiempo (Años)")
plt.ylabel("Energía (kg. de aceite equivalente per cápita)")
plt.show()
```

## Emision de dioxido de carbono dependiendo del uso de energia


```python
def polinomio_Energia_CO2(n):
    return np.polyfit(Energia,TonCO2,n)

#Vector para ordenar de manera puntual la Energia y poder construir las funciones. 
X=np.linspace(Energia.min(),Energia.max(),15)

p1=polinomio_Energia_CO2(1)
p2=polinomio_Energia_CO2(2)
p3=polinomio_Energia_CO2(3)

curva1=p1[0]*X**(1)+p1[1]*X**(0)
curva2=p2[0]*X**(2)+p2[1]*X**(1)+p2[2]*X**(0)
curva3=p3[0]*X**(3)+p3[1]*X**(2)+p3[2]*X**(1)+p3[3]*X**(0)



plt.plot(Energia, TonCO2, 'ob',label='Emisión CO2')
plt.plot(X,curva1,label="Ajuste primer orden")
plt.plot(X,curva2,label="Ajuste segundo orden")
plt.plot(X,curva3,label="Ajuste tercer orden")
plt.legend(loc='best')
plt.xlabel("Energía (kg. de aceite equivalente per cápita)")
plt.ylabel("Dióxido de Carbono (Toneladas)")
plt.title("Emision de dioxido de carbono dependiendo del uso de energia")
plt.show()
```
